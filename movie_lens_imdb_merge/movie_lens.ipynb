# movielens "small" dataset 

```python
# import various moviedata from movielens "small" dataset
# Small: 100,000 ratings and 3,600 tag applications applied to 9,000 movies by 600 users. Last updated 9/2018.
# https://grouplens.org/datasets/movielens/
```

```python
import pandas as pd
import numpy as np
```

```python


# mlens_links: 3 rows with IDs for internal, imdb and themoviedb 
# 1st row: "intenral" movie-number = number used in data set to address a movie over various files 
# 2nd row: imdb number: this number has to be completed with "[tt0]number" to work as an url: 116260 --> https://www.imdb.com/de/title/tt0116260/
# 3rd row: tmdb number: can be used directly in URLs, like 17182 as in https://www.themoviedb.org/movie/17182-eye-for-an-eye 
mlens_links = pd.read_csv("../input/movielens/links.csv")  #, index_col = "movieId")
mlens_movies = pd.read_csv("../input/movielens/movies.csv")  #, index_col = "movieId")
mlens_ratings = pd.read_csv("../input/movielens/ratings.csv")  #, index_col = "movieId")
mlens_tags = pd.read_csv("../input/movielens/tags.csv")  #, index_col = "movieId")
```

```python
# import imdb Data for comparison and clean up 
imdb_df = pd.read_parquet("../input/parquet_files/movies.parquet")

# clean up ../input/parquet_files/movies.parque
imdb_df = imdb_df.replace("", np.nan)

# Update d-type ratingValue
imdb_df["ratingValue"] = imdb_df["ratingValue"].astype(float)

# Test der Daten
imdb_df.sample(30)
```

# Join der Dateien

## Als Join Schlüssel wird die "interne" ID des Datasets [movieId] verwendet
Der Datensatz "Tags" wird vorerst nicht hinzugefügt

```python
merged_dataset_1 = pd.merge(mlens_links, mlens_movies)
merged_dataset = pd.merge(merged_dataset_1, mlens_ratings, how="outer")
```

```python
# Test des Merged data Set
#merged_dataset_1.sample(20)
merged_dataset.sample(20)
```

## Aufbereiten:
1. IMDB ID korrigieren
2. Spalte "movieId" entfernen
3. Spaltennamen korrigieren 
4. Datensatz exportieren als als Parquet-Datei

```python
# Status Data Set
merged_dataset.sample(30)
```

```python
# IMDB ID korrigieren
#merged_dataset["imdbId"].min()
#merged_dataset["imdbId"].max()

# Test: Filmdaten kleinster Wert 
#row = merged_dataset["imdbId"] == 417
# merged_dataset[row]

def rename(imdbid): 
    if len(str(imdbid)) == 7:
      return ("tt" + str(imdbid))
    
    elif len(str(imdbid)) < 7:
        new_imdbid = str(imdbid).zfill(7)
        return("tt" + new_imdbid)

merged_dataset["imdbId"] = merged_dataset["imdbId"].apply(lambda x : rename(x))
```

```python
# Test
merged_dataset.sort_values(by = "imdbId")
```

```python
# Angleichen der Spaltennamen
merged_dataset = merged_dataset.rename(columns = { "imdbId" : "ImdbId", "name" : "title", "genres" : "genre", "rating" : "ratingValue"})
```

```python
# Angleichen der Daten in Spalte Genre
merged_dataset["genre"] = merged_dataset["genre"].str.replace("|",", ")
```

```python
# Löschen der Spalte movieId
merged_dataset = merged_dataset.drop(columns=["movieId"])
```
